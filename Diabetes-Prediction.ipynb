```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
dia_df = pd.read_csv(r'C:\Users\NIMA\Downloads\diabetes.csv')
dia_df.head()
```

```python
dia_df.hist(bins=10,figsize=(10,10))
```

```python
sns.heatmap(dia_df.corr())
```

```python
sns.countplot(y=dia_df['Outcome'],palette='Set1')
```

```python
sns.set(style="ticks")
sns.pairplot(dia_df, hue="Outcome")
sns.set(style="whitegrid")
dia_df.boxplot(figsize=(15,6))
```

```python
sns.set(style="whitegrid")

sns.set(rc={'figure.figsize':(4,2)})
sns.boxplot(x=dia_df['Insulin'])
plt.show()
sns.boxplot(x=dia_df['BloodPressure'])
plt.show()
sns.boxplot(x=dia_df['DiabetesPedigreeFunction'])
plt.show()
```

```python
Q1=dia_df.quantile(0.25)
Q3=dia_df.quantile(0.75)
IQR=Q3-Q1

print("---Q1--- \n",Q1)
print("\n---Q3--- \n",Q3)
print("\n---IQR---\n",IQR)
```

The provided data displays the first quartile (Q1), third quartile (Q3), and interquartile range (IQR) for each column in the diabetes dataset. This information aids in identifying data distribution characteristics and potential outliers in features related to diabetes risk.
